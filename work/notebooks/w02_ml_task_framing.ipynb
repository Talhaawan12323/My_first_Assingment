## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
My lane is Lane 4: CTR / Engagement Opportunity Scoring.
As an ML task, this is a Ranking and Scoring problem specifically ranking candidates for review which can also be framed as an Anomaly Detection task finding pages whose CTR is anomalously low for their ranking position
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
My target is a CTR Gap Score. It represents how far below a pages CTR is compared to the expected CTR for its current position tier on Google. The wider the negative gap the higher the page climbs on the priority list.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
The success metric is Precision@K e.g Precision@50. If my model outputs the top 50 pages to fix how many of those 50 genuinely have strong visibility but poor CTR? We care about the top of the list because a human reviewer has limited time and can only process a few dozen pages
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")
visible_pages = df[(df['avg_position'] > 0) & (df['avg_position'] <= 10)]
expected_ctr = visible_pages.groupby('position_tier')['ctr'].transform('median')
visible_pages = visible_pages.copy()
visible_pages['ctr_gap'] = visible_pages['ctr'] - expected_ctr
cols_to_show = ['content_id', 'client_id', 'avg_position', 'ctr', 'ctr_gap']
print("Unit of Analysis: One row equals one unique page.")
print(visible_pages[cols_to_show].head())
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
A fixed rule would say something like: "Fix pages with CTR < 2%". But that is too rigid. CTR naturally changes based on intent Informational vs Transactional and exact position Rank 1 gets more clicks than Rank 9.
```
